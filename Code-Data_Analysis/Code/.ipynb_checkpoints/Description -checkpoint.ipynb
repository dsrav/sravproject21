```python
import pandas as pd

numeric_df = df.select_dtypes(include='number')  # Select only numeric columns

# Calculating mean, median, and standard deviation
mean_values = numeric_df.mean()
median_values = numeric_df.median()
std_dev_values = numeric_df.std()

# Displaying the results
print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nStandard deviation values:\n", std_dev_values)

```

```python
import pandas as pd

# Calculating the mean for the columns with missing values
mean_avg_lat_down_ms = df['avg_lat_down_ms'].mean()
mean_avg_lat_up_ms = df['avg_lat_up_ms'].mean()

# Imputing missing values with the mean
df['avg_lat_down_ms'].fillna(mean_avg_lat_down_ms, inplace=True)
df['avg_lat_up_ms'].fillna(mean_avg_lat_up_ms, inplace=True)

# Verifying that missing values have been filled
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
# Verifying that missing values have been filled
print("Missing values after imputation:")
print(df[['avg_lat_down_ms', 'avg_lat_up_ms']].isnull().sum())
```

```python
import pandas as pd
from scipy.stats import zscore

columns_to_check = ['avg_d_kbps', 'avg_u_kbps', 'avg_lat_ms']

# z-scores for specified columns
z_scores = zscore(df[columns_to_check])

# threshold for identifying outliers (e.g., 3 standard deviations)
threshold = 3

# outliers based on the threshold
outliers = (abs(z_scores) > threshold).any(axis=1)

# Displaying the rows with outliers
outliers_df = df[outliers]
print("Rows with outliers:")
print(outliers_df)

# Removing outliers from the original DataFrame
df_cleaned = df[~outliers]

# saving the cleaned DataFrame to a new file
df_cleaned.to_csv('cleaned_dataset.csv', index=False)
```
