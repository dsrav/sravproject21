```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\dsrav\Documents\2Uog files\data analytics\Code\rounded_cleaned_dataset.csv'
df = pd.read_csv(file_path)
```

```python
import numpy as np

# Select only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Display basic statistics
basic_statistics = numerical_df.describe()

# Calculate mode
mode_values = numerical_df.mode().iloc[0]

# Calculate correlation matrix
correlation_matrix = numerical_df.corr()

# Calculate variance and covariance
variance = numerical_df.var()
covariance_matrix = numerical_df.cov()

# Display the results
print("Basic Statistics:")
print(basic_statistics)

print("\nMode Values:")
print(mode_values)

print("\nCorrelation Matrix:")
print(correlation_matrix)

print("\nVariance:")
print(variance)

print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
descriptive_stats = numerical_df.describe()

#  display the results
display(descriptive_stats)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize Basic Statistics
plt.figure(figsize=(12, 6))
basic_statistics.drop('count').plot(kind='bar', ax=plt.gca())
plt.title('Basic Statistics')
plt.xticks(rotation=0)
plt.show()

# Visualize Mode Values
plt.figure(figsize=(12, 6))
mode_values.T.plot(kind='bar', ax=plt.gca())
plt.title('Mode Values')
plt.xticks(rotation=0)
plt.show()

# Visualize Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Visualize Variance
plt.figure(figsize=(8, 6))
variance.plot(kind='bar', ax=plt.gca())
plt.title('Variance')
plt.xticks(rotation=0)
plt.show()

# Visualize Covariance Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Covariance Matrix')
plt.show()
```
